```python
from myPyPaperBot.PyPaperBot import __main__ as ppbot
```

```python
no_link_files = 'parsing-output/pdf_names_wo_link.txt'
```

```python
no_link_names = []
no_link_queries = []
with open(no_link_files) as file_in:
    for line in file_in:
        if line[-1] == '\n':
            split_line = line[:-1].split(',', 1)
            no_link_names.append(split_line[0])
            no_link_queries.append(split_line[1])
        else:
            split_line = line.split(',', 1)
            no_link_names.append(split_line[0])
            no_link_queries.append(split_line[1])
```

```python
dwn_dir = 'test-papers/'
scihub_mirror = 'https://sci-hub.se/'
scholar_pages = range(1)
```

```python
ppbot.start(query='', scholar_results=10, proxy=None, scholar_pages=scholar_pages, dwn_dir=dwn_dir, 
            SciHub_URL=scihub_mirror, queries=no_link_queries[0:1], query_pdf_names=no_link_names[0:1])
```

```python
ppbot.start(no_link_queries[0], scholar_results=10, proxy=None, scholar_pages=scholar_pages, dwn_dir=dwn_dir, 
            SciHub_URL=scihub_mirror)
```

```python
for i, (name, title) in enumerate(zip(no_link_names, no_link_queries)):
    print(i ,name, title)
```
